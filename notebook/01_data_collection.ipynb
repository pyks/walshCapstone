- `--pages 1900` → full dataset (~22,450 records, takes several hours)

## 1.2 Load the Raw Dataset

```python
RAW_PATH = Path("../data/raw/fragrantica_raw.csv")

df_raw = pd.read_csv(RAW_PATH)
print(f"Shape: {df_raw.shape}")
df_raw.head()
```

```python
df_raw.info()
```

## 1.3 Download Secondary Dataset (Parfumo / TidyTuesday)

```python
PARFUMO_URL = (
    "https://raw.githubusercontent.com/rfordatascience/tidytuesday/"
    "main/data/2024/2024-12-10/parfumo_data_clean.csv"
)

df_parfumo = pd.read_csv(PARFUMO_URL)
print(f"Parfumo dataset shape: {df_parfumo.shape}")
df_parfumo.head()
```

```python
# Save Parfumo data to raw folder
df_parfumo.to_csv("../data/raw/parfumo_raw.csv", index=False)
print("Parfumo dataset saved.")
```
